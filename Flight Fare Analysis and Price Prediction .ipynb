## Stage 2 – Data Cleaning and Pre-processing

Handle missing values (impute or drop)

Handle duplicates

Treat outliers if required

Check skewness and apply transformations

Convert data types if needed

Feature transformations (date parts, derived fields if required for analysis)

**Handle missing values**

```python
# Handling Missing Values
print(df.isnull().sum())
```

```python
# Fill missing Route and Total Stops by mode

df['Route'].fillna(df['Route'].mode()[0], inplace=True)
df['Total_Stops'].fillna(df['Total_Stops'].mode()[0], inplace=True)
```

```python
# Verifying Missing Values

print(df.isnull().sum())
```

**Handle Duplicates**

```python
# remove duplicates

df.drop_duplicates(inplace=True)

df.duplicated().sum()
```

**Convert Data Types**

```python
# Convert Date_of_Journey to datetime

df['Date_of_Journey'] = pd.to_datetime(df['Date_of_Journey'], format='%d/%m/%Y')
```

```python
# Convert Total_Stops into numeric

df['Total_Stops'].unique()
```

```python
df['Total_Stops'] = df['Total_Stops'].fillna(df['Total_Stops'].mode()[0])

df['Total_Stops'] = df['Total_Stops'].replace({
    'non-stop': 0,
    '1 stop': 1,
    '2 stops': 2,
    '3 stops': 3,
    '4 stops': 4
})

df['Total_Stops'] = df['Total_Stops'].astype(int)

df
```

**Convert Duration into minutes**

```python
# Convert text duration into numeric format


def convert_duration(duration):
    h = 0
    m = 0
    if 'h' in duration:
        h = int(duration.split('h')[0])
    if 'm' in duration:
        m = int(duration.split('m')[0].split()[-1])
    return h * 60 + m

df['Duration_Minutes'] = df['Duration'].apply(convert_duration)
df.columns
```

**Date & Time Features**

```python
# Extract journey details

df['Date_of_Journey'] = pd.to_datetime(
    df['Date_of_Journey'],
    format='%d/%m/%Y',
    errors='coerce'
)

df['Journey_Day'] = df['Date_of_Journey'].dt.day
df['Journey_Month'] = df['Date_of_Journey'].dt.month
df['Journey_Weekday'] = df['Date_of_Journey'].dt.day_name()

df['Date_of_Journey'].dtype
```

```python
#verifying

df[['Date_of_Journey', 'Journey_Day', 'Journey_Month', 'Journey_Weekday']].head()
```

```python
# Extract departure hour

df['Dep_Hour'] = pd.to_datetime(
    df['Dep_Time'],
    format='%H:%M',
    errors='coerce'
).dt.hour

# verifying

df[['Dep_Time', 'Dep_Hour']].head()
```

```python
# Data Consistency Checks

df['Airline'].value_counts()
df['Source'].value_counts()
df['Destination'].value_counts()
```

```python
# Detect outliers using IQR

Q1 = df['Price'].quantile(0.25)
Q3 = df['Price'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
```

```python
# Remove extreme outliers

df = df[(df['Price'] >= lower) & (df['Price'] <= upper)]
```

```python
# Check skewness

df['Price'].skew()
```

```python
# Apply log transformation

df['Price_log'] = np.log1p(df['Price'])
```

# **Type of Analysis**

 **1. DESCRIPTIVE ANALYSIS (What happened?)**


* The dataset contains 1,680 flight records with an average price of ₹9.40K, a minimum price of ₹2,050, and a maximum price of ₹23K.

* Ticket prices vary across different days of the week, with higher average prices observed on Sunday and Wednesday.

* The average price increases as the number of stops increases, with the highest prices for flights having 4 stops.

* Monthly analysis shows that prices are lowest in Month 4 and higher in Months 3, 5, and 6.

* Route-based and source-destination analysis shows that certain routes and cities such as Bangalore, Kolkata, and BOM-based routes have higher average fares.




**2. DIAGNOSTIC ANALYSIS (Why did it happen?)**

* Prices are higher for flights with more stops because longer and more complex routes increase overall travel cost.

* The dip in prices during Month 4 suggests a period of lower travel demand compared to other months.

* Higher prices on Sunday and Wednesday indicate increased demand on these travel days.

* outes such as BOM–Delhi and BLR–Kolkata show higher prices due to route-specific demand patterns.

* The large gap between the maximum price (₹23K) and the average price (₹9.40K) indicates the presence of premium or peak-time bookings.

**3. PREDICTIVE ANALYSIS (What is likely to happen?)**

* Future flights with more stops are likely to have higher ticket prices compared to non-stop flights.

* Ticket prices are expected to be lower during Month 4 and higher in later months based on observed trends.

* Routes that currently show higher average prices are likely to continue being expensive in future bookings.

* Flights scheduled on Sunday and Wednesday are likely to remain costlier than those on other weekdays.

* If booking conditions remain similar, the expected average ticket price will stay around ₹9.40K.

**4. PRESCRIPTIVE ANALYSIS (What should be done?)**

* Travelers should prefer non-stop or one-stop flights to reduce ticket costs.

* Booking travel during Month 4 is recommended to obtain lower fares.

* Passengers should avoid high-priced routes and peak weekdays when cost savings are a priority.

* Airlines can use route and month-based trends to optimize pricing strategies.

* Travel platforms can use factors such as stops, route, and travel month to suggest cheaper flight options.
